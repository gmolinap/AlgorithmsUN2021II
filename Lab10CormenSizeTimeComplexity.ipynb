```python
import math

def nlogn(t):
    n = 2
    max = t*100*10**6
    nlogn = 0
    while(nlogn < max):
        n = max/math.log2(n)
        nlogn += (max-nlogn)/2
    return int(n)

print("Second: \n n <=", nlogn(1))
print("Minute: \n n <=", nlogn(60))
print("Hour: \n n <=", nlogn(3600))
print("Day: \n n <=", nlogn(86400))
print("Year: \n n <=", nlogn(31536000))
print("Century: \n n <=", nlogn(3153600000))
print("Milenium: \n n <=", nlogn(31536000000))
```

```python
import math

def nfac(t):
    n = 0
    max = t*100*10**6
    nf = 1
    while(nf <= max):
        n = n + 1
        nf = math.factorial(n)
    return int(n-1)

print("Second: \n n <=", nfac(1))
print("Minute: \n n <=", nfac(60))
print("Hour: \n n <=", nfac(3600))
print("Day: \n n <=", nfac(86400))
print("Year: \n n <=", nfac(31536000))
print("Century: \n n <=", nfac(3153600000))
print("Milenium: \n n <=", nfac(31536000000))
```

```python
import math

def nn(t):
    n = 0
    max = t*100*10**6
    nn = 1
    while(nn <= max):
        n = n + 1
        nn = n**n
    return int(n-1)

print("Second: \n n <=", nn(1))
print("Minute: \n n <=", nn(60))
print("Hour: \n n <=", nn(3600))
print("Day: \n n <=", nn(86400))
print("Year: \n n <=", nn(31536000))
print("Century: \n n <=", nn(3153600000))
print("Milenium: \n n <=", nn(31536000000))
```
